```python
#%pip install ipywidgets
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

mass_sun = 1.989e30   # kg
mass_earth = 5.97e24  # kg
mass_mars = 6.418e23

G = 6.67430e-11  # m^3 kg^-1 s^-2

# perihelion halley
AU = 149e9         #m

v_earth = 29.78e3   # m/s
r_earth = AU  # m 

v_mars = 24.07e3     #m/s
r_mars = 1.5*AU   #m

# settings
dt = 24 * 3600  # seconds
t = 687 * 24 * 3600  # seconds
N = int(t / dt)

# Ic.
v_e = np.zeros((N,2))
v_e[0,:] = [0,v_earth]

r_e = np.zeros((N,2))
r_e[0,:] = [r_earth,0]

v_m = np.zeros((N,2))
v_m[0,:] = [0,v_mars]

r_m = np.zeros((N,2))
r_m[0,:] = [r_mars,0]

t_arr = np.linspace(0,t+dt,N)
```

```python
# RK4 method 

def acceleration(r):
    r_norm = np.linalg.norm(r)
    return -G * mass_sun * r / r_norm**3

#earth
for i in range(1, N):
    # r and v at time t
    r0 = r_e[i-1, :]
    v0 = v_e[i-1, :]
    
    # k1
    k1_v = acceleration(r0) * dt
    k1_r = v0 * dt

    # k2
    k2_v = acceleration(r0 + 0.5 * k1_r) * dt
    k2_r = (v0 + 0.5 * k1_v) * dt

    # k3
    k3_v = acceleration(r0 + 0.5 * k2_r) * dt
    k3_r = (v0 + 0.5 * k2_v) * dt

    # k4
    k4_v = acceleration(r0 + k3_r) * dt
    k4_r = (v0 + k3_v) * dt

    # update velocity and position
    v_e[i, :] = v0 + (k1_v + 2 * k2_v + 2 * k3_v + k4_v) / 6
    r_e[i, :] = r0 + (k1_r + 2 * k2_r + 2 * k3_r + k4_r) / 6

#mars
for i in range(1, N):
    # r and v at time t
    r0 = r_m[i-1, :]
    v0 = v_m[i-1, :]
    
    # k1
    k1_v = acceleration(r0) * dt
    k1_r = v0 * dt

    # k2
    k2_v = acceleration(r0 + 0.5 * k1_r) * dt
    k2_r = (v0 + 0.5 * k1_v) * dt

    # k3
    k3_v = acceleration(r0 + 0.5 * k2_r) * dt
    k3_r = (v0 + 0.5 * k2_v) * dt

    # k4
    k4_v = acceleration(r0 + k3_r) * dt
    k4_r = (v0 + k3_v) * dt

    # update velocity and position
    v_m[i, :] = v0 + (k1_v + 2 * k2_v + 2 * k3_v + k4_v) / 6
    r_m[i, :] = r0 + (k1_r + 2 * k2_r + 2 * k3_r + k4_r) / 6

```

```python
# Plotting the trajectory
plt.figure()
plt.scatter(0, 0, color='yellow', s=100, label='Sun')  # Sun at origin
plt.plot(r_e[:, 0], r_e[:, 1], label='Earth Trajectory', color='lightgray')
plt.plot(r_m[:, 0], r_m[:, 1], label='Mars Trajectory', color='lightblue')
plt.show()
```

```python
def sim_kep(t_sim):

    plt.figure(figsize=(6, 6))
    plt.xlabel('X Position (m)')
    plt.ylabel('Y Position (m)')

    plt.plot(r_e[:, 0], r_e[:, 1], label='Earth Trajectory', color='lightgray')
    plt.plot(r_m[:, 0], r_m[:, 1], label='Mars Trajectory', color='lightgray')


    # plt.plot(r_h[N, 0], r_h[N, 1], 'r.')

    plt.scatter(0, 0, color='yellow', s=100, label='Sun')  # Sun at origin
    plt.gca().set_aspect('equal')
    plt.title('Earth & Mars Trajectory Around the Sun')


    for i in range(1,t_sim,int(N/15)):
        plt.plot(r_e[:i, 0], r_e[:i, 1], label='Earth Trajectory', color='black')
        plt.plot(r_m[:i, 0], r_m[:i, 1], label='Mars Trajectory', color='black')
        plt.plot(r_e[i, 0], r_e[i, 1], 'b.')
        plt.plot(r_m[i, 0], r_m[i, 1], 'r.')
        

    plt.text(0,1e11, str(int(t_sim))+'days')
    plt.grid()
    #plt.legend()
    plt.show()

interact(sim_kep, t_sim=widgets.IntSlider(min=0, max=N-1, step=int(N/15), value=1))
```
