```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Figure ch language of figs, vector
start = [0, 0]
end = [1.92, 1.92]

plt.figure()
plt.arrow(start[0], start[1], end[0], end[1], 
         head_width=0.1, head_length=0.1, fc='black', ec='black')

# Add text "P" near the end of the vector
plt.text(end[0] + 0.1, end[1], 'P', fontsize=12)

plt.text(.5*end[0] - 0.1, .5*end[1]+.1, r'$\vec{r}$', fontsize=12)

# Set limits and grid
plt.xlim(-0.5, 2.5)
plt.ylim(-0.5, 2.5)

plt.gca().set_aspect('equal')
plt.grid(True)

#plt.savefig('../ch1vector.png', dpi=300)
plt.show()
```
